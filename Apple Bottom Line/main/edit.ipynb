```python
# import libraries
import pandas as pd
import numpy as np
import csv
```

```python
# Read the csv file to get the Records and all variables
record_list = []
variables = set()

with open('bottom_line_Apple.csv', 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        if len(row) == 1:
            record_list.append(row[0])
        else:
            row = row[1:]
            for i in range(0,len(row),2):
                variables.add(row[i])

print(record_list)
print(variables)
```

```python
# convert record_list to panda Series
record = pd.Series(record_list, name='Record')
record
```

```python
# convert Series into dataframe
# initalize all variables into the dataframe
df = pd.DataFrame(record, columns=['Record'])

variables = sorted(variables)

for variable in variables:
    df[variable] = [''] * len(df)

df
```

```python
# read the csv file again and fill the dataframe

dict = {}
current_record = ''

# Function to add values to the dataframe based on the current record
def add_to_df(current_record, df, dict):
    if dict.keys() == []:  # if dict is empty, do not proceed
        return
    if current_record in df['Record'].values:
        for variable, value in dict.items():
            df.loc[df['Record'] == current_record, variable] = value


with open('bottom_line_Apple.csv', 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        if len(row) == 1: # if the row has only one element, it is a record name
            if current_record != '': # check if current_record is not empty
                add_to_df(current_record, df, dict)
                dict = {}
            current_record = row[0]
        else:
            row = row[1:] # remove the first element which is the record name or empty string
            for i in range(0, len(row), 2):
                variable = row[i]
                value = row[i + 1] if i + 1 < len(row) else ''
                dict[variable] = value
    add_to_df(current_record, df, dict)
```

```python
df.drop('', axis=1, inplace=True)
df
```

```python
# dataframe cleaning
df['Record'] = df['Record'].str.split(' ').str[1]
df
```

```python
# export cleaned csv
df.to_csv('bottom_line_Apple_cleaned.csv', index=False)
```
